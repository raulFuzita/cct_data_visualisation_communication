```python
path = str(r"..\..\..\..\data\cct_classes\DVC\week2\50_Startups.csv").replace('\\', '//')
```

```python
df = pd.read_csv(path)
```

1) LINE PLOT COMPARING R&D AND ADMINISTRATION

```python
rd = df['R&D Spend']
rd.head()
```

```python
adm = df['Administration']
adm.head()
```

```python
plt.plot(rd)
plt.plot(adm)
plt.xlabel('State')
plt.ylabel('Spend')
plt.title('R&D vs Administration')
plt.legend(['R&D', 'Administration'])
```

2) LINE PLOT COMPARING ADM AND MARKETING

```python
mkt = df['Marketing Spend']
mkt.head()
```

```python
plt.plot(adm)
plt.plot(mkt)
plt.xlabel('State')
plt.ylabel('Spend')
plt.title('Administration vs Marketing')
plt.legend(['Administration', 'Marketing'])
```

3) BAR CHART FOR PROFITS IN NEW YORK

```python
ny = df[df.State == 'New York'][['Profit']]
ny
```

```python
plt.bar(ny.index, ny.Profit)
plt.xlabel('New York')
plt.ylabel('Profit')
plt.title('PROFITS IN NEW YORK')
```

HOMEWORK: BAR CHART FOR CALIFORNIA AND FLORIDA PROFIT
